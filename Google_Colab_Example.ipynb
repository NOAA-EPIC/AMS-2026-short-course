```python
#@title Typical linux commands for directory to create and read/execute custom python

import os
print(os.listdir())

# Create a file and write to it
file_name = "exampleText.txt"

# Open the file in write mode ('w'). If the file doesn't exist, it will be created.
with open(file_name, 'w') as file:
    file.write("Hello, this is a new file created using Python!")

print(f"{file_name} has been created successfully.")
```

```python
# Read from the file to verify it is what we created
with open(file_name, 'r') as file:
    content = file.read()

print(f"Contents of {file_name}:")
print(content)
```

```python
#@title The Google Colab import to download a custom text file

from google.colab import files
files.download(file_name)
```
